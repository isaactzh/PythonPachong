```python
import pymongo
from string import punctuation
import charts
```

```python
client = pymongo.MongoClient('localhost', 27017)
```

```python
ganji = client['ganji']
```

```python
item_ganji = ganji['item_ganji']
```

```python
#整理数据
for i in item_ganji.find().limit(300):
    print(i['area']) #只显示地区
```

```python
for i in item_ganji.find():
    if i['area']:
        #area = [i for i in i['area] if i not in punctuation]
        area = i['area'].split('-')
    else:
        area = ['不明']
    item_ganji.update_many({'_id':i['_id']},{'$set':{'area':area}}) #选中所有的id
        
```

```python
#更新数据库中的值
#db.collection.update(源文件，修改结果)
#e.g. {id:1,name:0,info:3}
#update({id:1},{$set:{name:2}}) :更新对象，如果不存在，会创造新的对象
#$set:操作符
for i in item_ganji.find().limit(300):
    print(i['area'])
```

```python
#数据整理，名称对应的次数：
#将所有地区的列表去重：
area_list = []
for i in item_ganji.find():
    if(i['area'][0] == '北京'):
        area_list.append(i['area'][-1])
    area_index = list(set(area_list))
print(area_index)
```

```python
#创建统计量的列表：
post_times = []
for index in area_index:
    post_times.append(area_list.count(index))
print(post_times)
```

```python
#创建函数，生成结构：
def data_gen(types):
    length = 0
    if length <= len(area_index):
        for area, times in zip(area_index,post_times):
            data = {
                'name': area,
                'data': [times],
                'type': types
            }
            yield data
            length += 1
```

```python
data_gen('column')
```

```python
for i in data_gen('column'):
    print(i)
```

```python
series = [data for data in data_gen('column')]
charts.plot(series, show = 'inline', options=dict(title = dict(text='北京二手物品发帖量')))
```

```python
for i in item_ganji.find().limit(100):
    print(i['cate'])
```

```python
cate_list = []
for i in item_ganji.find():
    cate_list.append(i['cate'])
    cate_index = list(set(cate_list))
print(cate_index)
```

```python
cate_count = []
for index in cate_index:
    cate_count.append(cate_list.count(index))
print(cate_count)
```

```python
def cate_data_gen(types):
    length = 0
    if length <= len(cate_index):
        for cate, count in zip(cate_index, cate_count):
            data = {
                'name': cate,
                'data': [count],
                'type': types
            }
            yield data
            length += 1
```

```python
for data in cate_data_gen('column'):
    print(data)
```

```python
series = [data for data in cate_data_gen('column')]
charts.plot(series, show = 'inline', options=dict(title = dict(text='北京二手物品种类统计')))
```
